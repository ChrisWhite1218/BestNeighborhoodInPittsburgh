```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python

transport_2010 = pd.read_csv("transportation.csv")

Commute_2010 = pd.read_csv("transportation.csv", usecols=['Neighborhood','Population (2010)','Commute to Work: Drive Alone (2010)','Commute to Work: Carpool/Vanpool (2010)','Commute to Work: Public Transportation (2010)','Commute to Work: Taxi (2010)',
                                                         'Commute to Work: Motorcycle (2010)','Commute to Work: Bicycle (2010)','Commute to Work: Walk (2010)','Commute to Work: Other (2010)'])
Commute_2010 = Commute_2010.dropna()
# Makes the column named Neighborhood as the row index
# Creates the Commute_2010 and pulls in all the commute data from the csv


Commute_2010['Commute to Work: Drive Alone (2010)'] = Commute_2010['Commute to Work: Drive Alone (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Carpool/Vanpool (2010)'] = Commute_2010['Commute to Work: Carpool/Vanpool (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Public Transportation (2010)'] = Commute_2010['Commute to Work: Public Transportation (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Taxi (2010)'] = Commute_2010['Commute to Work: Taxi (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Motorcycle (2010)'] = Commute_2010['Commute to Work: Motorcycle (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Bicycle (2010)'] = Commute_2010['Commute to Work: Bicycle (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Walk (2010)'] = Commute_2010['Commute to Work: Walk (2010)'].str.rstrip('%').astype('float')
Commute_2010['Commute to Work: Other (2010)'] = Commute_2010['Commute to Work: Other (2010)'].str.rstrip('%').astype('float')


```

The Above is reading in the CSV file of Transportation from 2010 on the Neighborhoods of Pittsburgh. And then taking the Commute values and changing them from a percentage, into something that can be plotted easily. This is being done by removeing the "%" and then changing the values to floats.


---------------------------------------------------------------------------------------------------------------------------------------------
The Below code is using the Commute_2010, and sorting the values in decenting order, based on the largest y data, which one average is "Commute to Work: Drive Alone (2010)". This looks nicer and is much easier to read then to sort by the names of the Neighborhoods.


```python
#del transport_2010['Street Density (st. mi/area sq. mi)']

#Commute_2010.head(10)
SortedTransport = Commute_2010.sort_values(by=["Commute to Work: Drive Alone (2010)"],ascending=False)

#Data = SortedTransport.plot.barh(x="Neighborhood",y="Commute to Work: Drive Alone (2010)",figsize =(10,30) )
SortedTransport.plot.barh(x='Neighborhood', y=['Commute to Work: Drive Alone (2010)','Commute to Work: Carpool/Vanpool (2010)',
                                                      'Commute to Work: Public Transportation (2010)','Commute to Work: Taxi (2010)',
                                                      'Commute to Work: Motorcycle (2010)','Commute to Work: Bicycle (2010)',
                                                     'Commute to Work: Walk (2010)','Commute to Work: Other (2010)'],figsize =(20,30),stacked=True)

plt.title("Commuting Per Neighborhood")
plt.xlabel("Percentage of different forms of Commuting Per Neighborhood")
plt.ylabel("Neighborhoods")
#Data.xLabel("Commute to work from Driving alone per Neighborhood")
#Data.plot(kind="bar")

#df = pd.DataFrame(Commute_2010)
```
